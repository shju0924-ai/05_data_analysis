다음과 같은 과일 판매량 데이터가 있습니다. 각 과일의 1분기와 2분기 판매량이 주어졌을 때, NumPy 배열을 사용하여 아래 문제를 해결하세요.

1. 1분기와 2분기의 과일별 총 판매량을 구하세요.
2. 과일별 평균 판매량을 구하세요.

- **데이터**:
    - 1분기 판매량: `[200, 340, 150, 400]` (사과, 배, 오렌지, 바나나)
    - 2분기 판매량: `[220, 300, 180, 420]` (사과, 배, 오렌지, 바나나)

```python
q1 = np.array([200, 340, 150, 400])
q2 = np.array([220, 300, 180, 420])

total = q1+q2
print(total)

sales = np.array([q1, q2])
avg = np.mean(sales, axis=0)
print(avg)
```

다음 두 배열을 사용하여 각각의 요소별로 덧셈, 뺄셈, 곱셈, 나눗셈을 수행하세요.

arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("덧셈: ", arr1+arr2)
print("뺄셈: ", arr1-arr2)
print("곱셈: ", arr1*arr2)
print("나눗셈: ", arr1/arr2)
```
