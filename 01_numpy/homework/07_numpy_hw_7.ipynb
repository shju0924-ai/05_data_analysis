다음은 5개 지점에서 판매된 과일에 대한 월별 판매량 데이터입니다. 이 데이터는 각 지점의 판매량을 월별로 나타내며, 각 지점에서 판매된 사과, 바나나, 오렌지의 판매량이 기록되어 있습니다. 이 데이터를 사용하여 여러 가지 통계적 분석과 배열 조작을 수행하세요.

1. 각 지점에서 6개월 동안 판매된 **사과**, **바나나**, **오렌지**의 월별 평균 판매량을 계산하세요.
2. 각 월마다 전체 지점에서 가장 많이 판매된 과일을 찾아 판매량과 과일명을 출력하세요.
3. 6개월 동안 전체 지점에서 가장 많이 판매된 과일의 총량을 과일별로 구하세요. 어떤 과일이 가장 많이 판매되었는지 확인하세요.
4. 각 지점에서 판매량이 150개 이상인 데이터를 추출하고 그에 해당하는 지점과 과일을 출력하세요.

- **데이터 (**각 지점의 월별 과일 판매량(단위: 개))

    ```python
    import numpy as np

    # 5개 지점의 월별 사과, 바나나, 오렌지 판매량 (5x3x6 배열)
    # 5개의 지점, 3개의 과일(사과, 바나나, 오렌지), 6개월치 데이터
    sales_data = np.array([
        # 지점 1
        [[120, 150, 200, 230, 180, 210],  # 사과 판매량
         [100, 140, 170, 190, 160, 180],  # 바나나 판매량
         [90, 120, 160, 140, 130, 150]],  # 오렌지 판매량

        # 지점 2
        [[80, 90, 110, 140, 130, 120],
         [60, 70, 90, 100, 110, 95],
         [50, 80, 100, 110, 90, 85]],

        # 지점 3
        [[130, 140, 160, 180, 170, 190],
         [110, 120, 130, 150, 140, 160],
         [100, 110, 120, 130, 120, 140]],

        # 지점 4
        [[100, 130, 150, 170, 160, 180],
         [90, 100, 110, 120, 130, 125],
         [80, 110, 140, 130, 120, 110]],

        # 지점 5
        [[90, 110, 130, 140, 120, 130],
         [70, 85, 100, 110, 105, 95],
         [60, 90, 110, 100, 90, 95]]
    ])
    ```

````python
sales_data = np.array([
    # 지점 1
    [[120, 150, 200, 230, 180, 210],
     [100, 140, 170, 190, 160, 180],
     [90,  120, 160, 140, 130, 150]],
    # 지점 2
    [[80,  90,  110, 140, 130, 120],
     [60,  70,  90,  100, 110,  95],
     [50,  80,  100, 110,  90,  85]],
    # 지점 3
    [[130, 140, 160, 180, 170, 190],
     [110, 120, 130, 150, 140, 160],
     [100, 110, 120, 130, 120, 140]],
    # 지점 4
    [[100, 130, 150, 170, 160, 180],
     [90,  100, 110, 120, 130, 125],
     [80,  110, 140, 130, 120, 110]],
    # 지점 5
    [[90,  110, 130, 140, 120, 130],
     [70,   85, 100, 110, 105,  95],
     [60,   90, 110, 100,  90,  95]]
])

fruits = ['사과', '바나나', '오렌지']
branches = ['지점1', '지점2', '지점3', '지점4', '지점5']

monthly_avg = np.mean(sales_data, axis=2)  # shape (5, 3)

for i, branch in enumerate(branches):
    print(f"{branch} → ", end="")
    for j, fruit in enumerate(fruits):
        print(f"{fruit}: {monthly_avg[i, j]:.1f}", end="  ")
    print()

monthly_total = np.sum(sales_data, axis=0)
best_fruit_idx = np.argmax(monthly_total, axis=0)

for m in range(6):
    idx = best_fruit_idx[m]
    qty = monthly_total[idx, m]
    print(f"{m+1}월 → {fruits[idx]} ({qty}개)")

total_by_fruit = np.sum(sales_data, axis=(0, 2))
best_total_idx = np.argmax(total_by_fruit)

for j, fruit in enumerate(fruits):
    print(f"{fruit}: {total_by_fruit[j]}개")
print(f"→ 가장 많이 팔린 과일: {fruits[best_total_idx]} ({total_by_fruit[best_total_idx]}개)")

b_idx, f_idx, m_idx = np.where(sales_data >= 150)

for b, f, m in zip(b_idx, f_idx, m_idx):
    print(f"{branches[b]} | {fruits[f]} | {m+1}월 → {sales_data[b, f, m]}개")

````
